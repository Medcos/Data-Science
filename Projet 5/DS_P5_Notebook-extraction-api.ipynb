```python
import requests

# Définir l'URL de l'API à laquelle on souhaite effectuer la requête GET
url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

# Définir les paramètres de requête qui seront ajoutés à l'URL
querystring = { 'ingr': 'champagne', 'nutrition-type': 'cooking','category[0]': 'generic-foods'}

# Définir les paramètres de requête qui seront ajoutés à l'URL
headers = {
	"X-RapidAPI-Key": "d85434e13fmsh2f4149e8f900a5cp1111e9jsn9fe40fac2404",  # spécifie la clé d'API
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"     # spécifie l'hôte d'API
}

#  Effectue la requête GET
response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

Le format JSON (JavaScript Object Notation) est un format de données léger et largement utilisé pour échanger des données entre un serveur et un client, et pour stocker des données structurées.

Le JSON est principalement composé de deux structures de données :

+ **Objet JSON** : Un objet JSON est une collection non ordonnée de paires clé-valeur, où les clés sont des chaînes de caractères et les valeurs peuvent être de différents types de données (chaînes de caractères, nombres, booléens, objets JSON imbriqués, tableaux JSON, etc.). Les paires clé-valeur sont séparées par des virgules et entourées par des accolades {}. 

+ **Tableau JSON** : Un tableau JSON est une liste ordonnée de valeurs, où chaque valeur peut être de différents types de données. Les valeurs sont séparées par des virgules et entourées par des crochets []

# <a name="C3"> III.) EXTRACTION DANS UN FICHIER CSV <a/>

```python
# Stocker la valeur associée à la clé "hints" dans une variable
dd = data.get("hints")
dd
```

```python
data_food = list(map(lambda x : x["food"], dd))
```

Voici une explication détaillée :

+ **dd** : C'est la liste contenant des éléments, qui sont des dictionnaires.
+ **lambda x: x["food"]** : C'est une fonction lambda qui prend un dictionnaire x en argument et retourne la valeur associée à la clé "food" dans ce dictionnaire.
+ **map(lambda x: x["food"], dd)** : C'est une fonction map() qui applique la fonction lambda à chaque élément de la liste dd. Elle renvoie un objet de type map qui contient les résultats de la transformation.
+ **list(map(lambda x: x["food"], dd))** : C'est une expression qui convertit l'objet map en une liste à l'aide de la fonction list(). Les résultats de la transformation sont maintenant stockés dans la liste data_food.

```python
food = pd.DataFrame(data_food)
food.head()
```

```python
food.shape
```

```python
food = food.loc[:, ['foodId', 'label','category', 'foodContentsLabel', 'image']]
food.head()
```

```python
foods=food.loc[food['image'].notnull(),:]
foods
```

# <a name="C4"> IV.) AFFICHAGE <a/>

```python
import pandas as pd
from PIL import Image
import requests
from io import BytesIO

# Créer nue liste vide
listes = []
# Parcourir la dataframe et afficher les images
for link in foods['image']:
    response = requests.get(link) #  envoie une requête GET à l'URL spécifiée par link pour récupérer les données de l'image
    img = Image.open(BytesIO(response.content)) # ouvrir l'image à partir des données de réponse
    listes.append(img) # Ajouter l'image à la listes
```

```python
import matplotlib.pyplot as plt

# Afficher les images
for i in range(len(listes)):
    plt.figure()
    plt.imshow(listes[i])
    plt.axis('ON')  # Afficher les axes
    plt.show()
```
