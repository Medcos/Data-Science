## 2.2 Tri par indicateurs 
### 2.2.1 Tri décroissant chacun des cinq (5) indicateurs

```python
# Gross enrolment ratio, tertiary, both sexes (%)

data5_1 = data5.loc[data5['Indicator Name'] == 'Gross enrolment ratio, tertiary, both sexes (%)']
data5_1 = data5_1.sort_values('2015', ascending = False)
data5_1 = data5_1.iloc[:10]
data5_1 = data5_1['Country Name'].tolist()
```

```python
# Gross enrolment ratio, upper secondary, both sexes (%)

data5_2 = data5.loc[data5['Indicator Name'] == 'Gross enrolment ratio, upper secondary, both sexes (%)']
data5_2 = data5_2.sort_values('2015', ascending = False)
data5_2 = data5_2.iloc[:10]
data5_2 = data5_2['Country Name'].tolist()
```

```python
# Internet users (per 100 people)

data5_3 = data5.loc[data5['Indicator Name'] == 'Internet users (per 100 people)']
data5_3 = data5_3.sort_values('2015', ascending = False)
data5_3 = data5_3.iloc[:10]
data5_3 = data5_3['Country Name'].tolist()
```

```python
# Population of the official age for tertiary education, both sexes (number)

data5_4 = data5.loc[data5['Indicator Name'] == 'Population of the official age for tertiary education, both sexes (number)']
data5_4 = data5_4.sort_values('2015', ascending = False)
data5_4 = data5_4.iloc[:10]
data5_4 = data5_4['Country Name'].tolist()
```

```python
# Population of the official age for upper secondary education, both sexes (number)

data5_5 = data5.loc[data5['Indicator Name'] == 'Population of the official age for upper secondary education, both sexes (number)']
data5_5 = data5_5.sort_values('2015', ascending = False)
data5_5 = data5_5.iloc[:10]
data5_5 = data5_5['Country Name'].tolist()
```

### 2.2.2 Liste des cinq pays 

```python
# Fusionner les listes en une liste
Data = data5_1+data5_2+data5_3+data5_4+data5_5

# Convertir la nouvelle liste en DataFrame
Data = pd.DataFrame(Data, columns=['Pays'])

# Compter le nbre d'occurences de chaque pays
Data =Data['Pays'].value_counts()

# Afficher les cinq premiers éléments
Data =Data.iloc[:5]

Data
```

```python
# Convertir l'index en liste
Liste = list(Data.index)
Liste
```

```python
DATA = data5[data5['Country Name'].isin(Liste)]
DATA
```
